```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([0,6])
ypoints=np.array([0,96])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
plt.plot(y1)
plt.xlabel("Avarage Pulse")
plt.ylabel("calories burnage")
plt.show
```

```python
plt.plot(y1)
plt.title("Test data")
plt.xlabel("Avarage Pulse")
plt.ylabel("calories burnage")
plt.show
```

```python
plt.plot(y1)
font1={"family":'serif',"color":"k","size":20}
font2={"family":'serif',"color":"red","size":25}
plt.title("Test Data",font2)
plt.xlabel("Avarage Pulse",fontdict=font1)
plt.ylabel("calories burnage",fontdict=font1)
plt.show
```

```python
plt.plot(y1)
font1={"family":'serif',"color":"k","size":20}
font2={"family":'serif',"color":"red","size":25}
plt.title("Test Data",font2,loc='left')
plt.xlabel("Avarage Pulse",fontdict=font1)
plt.ylabel("calories burnage",fontdict=font1)
plt.show
```

```python
plt.plot(y1)
font1={"family":'serif',"color":"k","size":20}
font2={"family":'serif',"color":"red","size":25}
plt.title("Test Data",font2)
plt.xlabel("Avarage Pulse",fontdict=font1)
plt.ylabel("calories burnage",fontdict=font1)
plt.grid()
plt.show
```

```python
plt.plot(y1)
font1={"family":'serif',"color":"k","size":20}
font2={"family":'serif',"color":"red","size":25}
plt.title("Test Data",font2)
plt.xlabel("Avarage Pulse",fontdict=font1)
plt.ylabel("calories burnage",fontdict=font1)
plt.grid(axis='y')
plt.show
```

```python
plt.plot(y1)
font1={"family":'serif',"color":"k","size":20}
font2={"family":'serif',"color":"red","size":25}
plt.title("Test Data",font2)
plt.xlabel("Avarage Pulse",fontdict=font1)
plt.ylabel("calories burnage",fontdict=font1)
plt.grid(axis='x')
plt.show
```

```python
plt.plot(y1)
font1={"family":'serif',"color":"k","size":20}
font2={"family":'serif',"color":"red","size":25}
plt.title("Test Data",font2)
plt.xlabel("Avarage Pulse",fontdict=font1)
plt.ylabel("calories burnage",fontdict=font1)
plt.grid(color='r',ls="-.",linewidth=0.5)
plt.show
```
